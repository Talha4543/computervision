```python
import cv2

# Paths to your images
image_paths = [
    "D:\Pics/download (1).jpg",  # Replace with the actual path to image 1
    "D:\Pics/download (2).jpg",  # Replace with the actual path to image 2
    "D:\Pics/download (3).jpg",  # Replace with the actual path to image 3
    "D:\Pics/download (4).jpg",
    "D:\Pics/download (5).jpg"  
]

# Load and display each image
for idx, image_path in enumerate(image_paths):
    # Load the image
    image = cv2.imread(image_path)

    # Check if the image was loaded successfully
    if image is not None:
        # Display the image in a window
        cv2.imshow(f'Image {idx + 1}', image)

        # Wait for a key press to move to the next image
        cv2.waitKey(0)
    else:
        print(f"Failed to load image at {image_path}")

# Close all windows after showing images
cv2.destroyAllWindows()
```
